```python
df_train = df_training[:]
df_test = null_data_fi[:]
Y_28 = df_training['mvar28']
Y_29 = df_training['mvar29']
Y_30 = df_training['mvar30']
```

```python
df_train = df_train.ix[:,2:29]
df_test = df_test.ix[:,2:29]
```

```python
X_Tr = df_train.to_dict('records')
X_Te = df_test.to_dict('records')
X_tr = []
X_te = []
X_tr.extend(X_Tr)
X_te.extend(X_Te)
X_total = X_tr + X_te
```

```python
#One Hot Encoding 
from sklearn.feature_extraction import DictVectorizer 
enc = DictVectorizer()
X_encoded_total =enc.fit_transform(X_total)
X_encoded_train =X_encoded_total[:len(X_tr)]
X_encoded_test =X_encoded_total[len(X_tr):]
```

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,ExtraTreesClassifier
from sklearn.linear_model import LogisticRegression
clf = RandomForestClassifier()
y_28 =  clf.fit(X_encoded_train.toarray(),Y_28).predict(X_encoded_test.toarray())
y_29 =  clf.fit(X_encoded_train.toarray(),Y_29).predict(X_encoded_test.toarray())
y_30 =  clf.fit(X_encoded_train.toarray(),Y_30).predict(X_encoded_test.toarray())
```

```python
null_data_fi['mvar28'] = y_28
null_data_fi['mvar29'] = y_29
null_data_fi['mvar30'] = y_30
```
